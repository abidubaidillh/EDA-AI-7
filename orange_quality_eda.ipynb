```python
import matplotlib.pyplot as plt
import seaborn as sns

def graph(y):
    sns.boxplot(x="Variety", y=y, data=df)
    plt.xticks(rotation=45)  # Memiringkan label jika kategori banyak
    plt.title(f"Boxplot of {y} by Variety")  # Menambahkan judul tiap grafik

# List kolom numerik yang ingin diplot
numerical_cols = ["Size (cm)", "Weight (g)", "Brix (Sweetness)", 
                  "Softness (1-5)", "HarvestTime (days)", 
                  "Ripeness (1-5)", "Quality (1-5)"]

# Menentukan jumlah baris dan kolom subplot
n_rows = (len(numerical_cols) // 3) + 1  # Maksimal 3 kolom per baris
fig, axes = plt.subplots(n_rows, 3, figsize=(20, n_rows * 5))

# Loop untuk membuat boxplot
for i, col in enumerate(numerical_cols):
    row, col_index = divmod(i, 3)  # Hitung posisi berdasarkan indeks
    plt.sca(axes[row, col_index])  # Pindah ke subplot yang sesuai
    graph(col)

# Menghapus subplot kosong agar tampilan lebih rapi
for j in range(i+1, n_rows * 3):  
    fig.delaxes(axes.flatten()[j])

plt.tight_layout()  # Menghindari tumpang tindih
plt.show()
```
