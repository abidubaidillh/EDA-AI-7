```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Orange Quality Data.csv")
```

```python
df.info()  # Melihat tipe data dan jumlah nilai yang hilang
df.describe()  # Statistik deskriptif
df.head()  # Menampilkan 5 baris pertama
```

```python
# get the shape of the dataset
df.shape
```

```python
# get info of the dataset
df.info()
```

```python
# get statistical summary
df.describe()
```

```python
# checking missing values
df.isnull().sum()
```

```python
# checking duplicate
data_nd = df.drop_duplicates(subset = "Variety")
data_nd
```

```python
# checking balance label
df.value_counts("Variety")
```

```python
# visualize each label
plt.figure(figsize=(12, 6))  # Perbesar ukuran figure
sns.countplot(x='Variety', data=df)
plt.xticks(rotation=45, ha="right")  # Rotasi label agar lebih jelas
plt.show()
```

```python
# Relation between variables
sns.scatterplot(x='Size (cm)', y='Weight (g)',
                hue='Variety', data=df, )
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```

```python
# Relation between variables
sns.scatterplot(x='', y='Weight (g)',
                hue='Variety', data=df, )
 
# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)
 
plt.show()
```
